```python
merged_columns = merged_df.columns.tolist()
print(merged_columns)
```

```python
import json

# Helper function using json.loads
def parse_json_col(col):
    try:
        return json.loads(col.replace("'", '"'))  # replace single quotes with double quotes for valid JSON
    except:
        return []

# Extract genres, director, and top 3 cast
clean_df = merged_df.copy()
clean_df['genres'] = clean_df['genres'].apply(lambda x: [d['name'] for d in parse_json_col(x)])
clean_df['director'] = clean_df['crew'].apply(lambda x: next((d['name'] for d in parse_json_col(x) if d.get('job')=='Director'), None))
clean_df['top_cast'] = clean_df['cast'].apply(lambda x: [d['name'] for d in parse_json_col(x)[:3]])

# Keep relevant columns
clean_df = clean_df[['title', 'budget', 'revenue', 'genres', 'director', 'top_cast', 
                     'popularity', 'vote_average', 'vote_count', 'release_date']]

# Preview
clean_df.head()
```

Handle missing values and datatypes

```python
clean_df.info()
```

```python
clean_df['release_date'] = pd.to_datetime(clean_df['release_date'])
print(clean_df['release_date'].dtype)
```

```python
clean_df = clean_df.dropna(subset=['director'])
clean_df = clean_df.dropna(subset=['release_date'])
```

conversion to categorical columns

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

clean_df['director_encoded'] = le.fit_transform(clean_df['director'])

clean_df.head()
```

```python
print(clean_df['director_encoded'].nunique())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

genres_encoded = pd.DataFrame(
    mlb.fit_transform(clean_df['genres']),
    columns=mlb.classes_,
    index=clean_df.index
)
```
