```python
import json

# Helper function using json.loads
def parse_json_col(col):
    try:
        return json.loads(col.replace("'", '"'))  # replace single quotes with double quotes for valid JSON
    except:
        return []

# Extract genres, director, and top 3 cast
clean_df = merged_df.copy()
clean_df['genres'] = clean_df['genres'].apply(lambda x: [d['name'] for d in parse_json_col(x)])
clean_df['director'] = clean_df['crew'].apply(lambda x: next((d['name'] for d in parse_json_col(x) if d.get('job')=='Director'), None))
clean_df['top_cast'] = clean_df['cast'].apply(lambda x: [d['name'] for d in parse_json_col(x)[:3]])

# Keep relevant columns
clean_df = clean_df[['title', 'budget', 'revenue', 'genres', 'director', 'top_cast', 
                     'popularity', 'vote_average', 'vote_count', 'release_date']]

# Preview
clean_df.head()
```

Handle missing values and datatypes

```python
clean_df.info()
```

```python
clean_df['release_date'] = pd.to_datetime(clean_df['release_date'])
print(clean_df['release_date'].dtype)
```

```python
clean_df = clean_df.dropna(subset=['director'])
clean_df = clean_df.dropna(subset=['release_date'])
```

```python
print(clean_df.info())
```

```python
clean_df.head()
```

conversion to categorical columns

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

clean_df['director_encoded'] = le.fit_transform(clean_df['director'])

clean_df.head()
```

```python
print(clean_df['director_encoded'].nunique())
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

genres_encoded = pd.DataFrame(
    mlb.fit_transform(clean_df['genres']),
    columns=mlb.classes_,
    index=clean_df.index
)
```

```python
genres_encoded.head()
```

```python
# 4️⃣ Extract year and month from release_date
# -----------------------------
clean_df['release_year'] = clean_df['release_date'].dt.year
clean_df['release_month'] = clean_df['release_date'].dt.month
```

```python
clean_df.head()
```

```python
import ast

# Convert string representations of lists to real Python lists
clean_df['top_cast'] = clean_df['top_cast'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Strip spaces and handle empty lists
clean_df['top_cast'] = clean_df['top_cast'].apply(
    lambda x: [actor.strip() for actor in x] if len(x) > 0 else ["Unknown"]
)

# Preview
clean_df['top_cast'].head(5)
```

```python
# Step 1: Calculate average rating per actor
actor_ratings = defaultdict(list)
for _, row in clean_df.iterrows():
    for actor in row['top_cast']:
        actor_ratings[actor].append(row['vote_average'])

actor_avg_rating = {actor: sum(r)/len(r) for actor, r in actor_ratings.items()}
```

```python
print(actor_avg_rating)
```

```python
def top_cast_avg_rating(top_cast):
    ratings = [actor_avg_rating.get(actor, 0) for actor in top_cast]  # use 0 if actor unseen
    return sum(ratings)/len(ratings) if ratings else 0

# Apply to dataset
clean_df['top_cast_avg_rating'] = clean_df['top_cast'].apply(top_cast_avg_rating)

# Preview
clean_df[['title', 'top_cast', 'top_cast_avg_rating']].head()
```
